```python
import numpy as np

def step_func(x):
    return 1 if x >= 0 else 0

def perceptron_learning(z, y, learning_rate=0.1, max_epochs=50):
    wei = np.zeros(z.shape[1])
    bias = 0
    for epoch in range(max_epochs):
        error_count = 0
        for i in range(len(z)):
            lin_op = np.dot(z[i], wei) + bias
            pred = step_func(lin_op)
            error = y[i] - pred
            if error != 0:
                error_count += 1
                wei += learning_rate * error * z[i]
                bias += learning_rate * error
        # Check if all predictions are correct
        predictions = [step_func(np.dot(z[i], wei) + bias) for i in range(len(z))]
        if np.array_equal(predictions, y):
            print(f"Training converged after epoch {epoch+1}")
            return wei, bias, epoch+1
    print("Did not converge within the given epochs")
    return wei, bias, max_epochs

# Training data for AND gate
z = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

wei, bias, epochs_taken = perceptron_learning(z, y)

def predict(z, wei, bias):
    lin_op = np.dot(z, wei) + bias
    return step_func(lin_op)

for x in z:
    print(f"Input: {x}, Predicted Output: {predict(x, wei, bias)}")

print(f"Result determined after epoch: {epochs_taken}")
```
