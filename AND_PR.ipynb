```python
import numpy as np

def step_func(x):
    return 1 if x>=0 else 0

def perceptron_learning(z,y,learning_rate=0.1,epochs=10):
    wei=np.zeros(z.shape[1])
    bias=0
    for epoch in range(epochs):
        for i in range(len(z)):
            lin_op=np.dot(z[i],wei)+bias
            pred=step_func(lin_op)
            error=y[i]-pred
            wei+=learning_rate*error*z[i]
            bias+=learning_rate*error
    return wei,bias
z=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
wei,bias=perceptron_learning(z,y)
    
def predict(z,wei,bias):
    lin_op=np.dot(z,wei)+bias
    return step_func(lin_op)
    
for x in z:
    print(f"Input: {x}, Predicted  Output: {predict(x,wei,bias)}")
```
